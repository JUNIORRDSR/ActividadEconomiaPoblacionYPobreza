```python
import pandas as pd

def clean_data_DCD(df):
    # Cambiar el nombre de la columna 'Unnamed: 0' a 'COD_DPTO'
    df = df.rename(columns={'Unnamed: 0': 'COD_DPTO'})
    # Cambiar el nombre de la columna 'Unnamed: 1' a 'DEPARTAMENTO'
    df = df.rename(columns={'Unnamed: 1': 'DEPARTAMENTO'})
    # Cambiar el nombre de la columna 'Unnamed: 2' a 'COD_DPTO-MPIO'
    df = df.rename(columns={'Unnamed: 2': 'COD_DPTO-MPIO'})
    # Cambiar el nombre de la columna 'Unnamed: 3' a 'MUNICIPIO'
    df = df.rename(columns={'Unnamed: 3': 'MUNICIPIO'})
    # Cambiar el nombre de la columna 'Unnamed: 4' a 'AÑO'
    df = df.rename(columns={'Unnamed: 4': 'AÑO'})
    # Cambiar el nombre de la columna 'Unnamed: 5' a 'ÁREA GEOGRÁFICA'
    df = df.rename(columns={'Unnamed: 5': 'ÁREA GEOGRÁFICA'})
    # Cambiar el nombre de la columna 'Unnamed: 6' a 'Total'
    df = df.rename(columns={'Unnamed: 6': 'Total'})
    # Cambiar el nombre de la columna 'Unnamed: 7' a 'Indígena'
    df = df.rename(columns={'Unnamed: 7': 'Indígena'})
    # Cambiar el nombre de la columna 'Unnamed: 8' a 'Gitano(a) o Rrom'
    df = df.rename(columns={'Unnamed: 8': 'Gitano(a) o Rrom'})
    # Cambiar el nombre de la columna 'Unnamed: 9' a 'Raizal del Archipiélago de San Andrés, Providencia y Santa Catalina'
    df = df.rename(columns={'Unnamed: 9': 'Raizal del Archipiélago de San Andrés, Providencia y Santa Catalina'})
    # Cambiar el nombre de la columna 'Unnamed: 10' a 'Palenquero(a) de San Basilio'
    df = df.rename(columns={'Unnamed: 10': 'Palenquero(a) de San Basilio'})
    # Cambiar el nombre de la columna 'Unnamed: 11' a 'Negro(a), mulato(a), afrodescendiente, afrocolombiano(a)'
    df = df.rename(columns={'Unnamed: 11': 'Negro(a), mulato(a), afrodescendiente, afrocolombiano(a)'})
    # Cambiar el nombre de la columna 'Unnamed: 12' a 'Ningún grupo étnico-racial'
    df = df.rename(columns={'Unnamed: 12': 'Ningún grupo étnico-racial'})
    # Drop rows from index 0 to 10
    df = df.drop(index=range(0, 1))
    return df

# Se ha cargado la variable "DCD_Proypoblacion_PerteneniaEtnicoRacialmun" del URI: c:\Users\junio\OneDrive\Escritorio\Inge\Universidad\5 Semestre\Fundamentos de economia\Actividad Población y Pobreza\anex-DCD-Proypoblacion-PerteneniaEtnicoRacialmun.xlsx
DCD_Proypoblacion_PerteneniaEtnicoRacialmun = pd.read_excel(r'c:\Users\junio\OneDrive\Escritorio\Inge\Universidad\5 Semestre\Fundamentos de economia\Actividad Población y Pobreza\anex-DCD-Proypoblacion-PerteneniaEtnicoRacialmun.xlsx')
DCD_clean = clean_data_DCD(DCD_Proypoblacion_PerteneniaEtnicoRacialmun.copy())
DCD_clean.head()
print(DCD_clean)
```
